```python
from collections import Counter

name = input("Enter a name: ")
vowels = "aeiouAEIOU" # Include both uppercase and lowercase vowels

# Count vowels using Counter and a generator expression
# Now, it checks against both upper and lower case vowels in the original name
vowel_counts = Counter(char for char in name if char in vowels)

output_parts = []
# Sort by vowel character for consistent output order
for vowel, count in sorted(vowel_counts.items()):
    output_parts.append(f"{vowel} {count} time{'s' if count > 1 else ''}")

print(", ".join(output_parts))
```

```python
name = input('Enter a name: ')
vowels = "aeiouAEIOU" # Include both uppercase and lowercase vowels
vowel_counts = {} # Use a dictionary to store counts

for char in name:
    if char in vowels:
        vowel_counts[char] = vowel_counts.get(char, 0) + 1

output_parts = [] # Use a list to build the output string
# Sort by vowel character for consistent output order
for vowel, count in sorted(vowel_counts.items()):
    output_parts.append(f"{vowel} {count} time{'s' if count > 1 else ''}")

print(", ".join(output_parts))
```

```python
def print_pattern(n):
    for i in range(1, n+1):
        # Print leading spaces
        print(' ' * (n - i), end='')

        # Print increasing sequence
        for j in range(1, i+1):
            print(j, end='')

        # Print decreasing sequence
        for k in range(i-1, 0, -1):
            print(k, end='')

        print()

# Example usage:
print_pattern(5)
```

```python
def factorial_recursive(n):
    if n < 0:
        raise ValueError("Factorial is not defined for negative numbers")
    elif n == 0 or n == 1:
        return 1
    else:
        return n * factorial_recursive(n-1)

# Example usage:
print(factorial_recursive(5))
```

```python
import math

def get_number():
    while True:
        try:
            num = int(input("Enter a non-negative integer: "))
            if num < 0:
                print("Factorial is not defined for negative numbers. Try again!")
            else:
                return num
        except ValueError:
            print("Invalid input. Please enter an integer.")

def factorial(n):
    return math.prod([i for i in range(1, n+1)])

def main():
    num = get_number()
    fact = factorial(num)
    print(f"The factorial of {num} is {fact}")

if __name__ == "__main__":
    main()
```

```python
def get_number():
    while True:
        try:
            num = int(input("Enter a non-negative integer: "))
            if num < 0:
                print("Factorial is not defined for negative numbers. Try again!")
            else:
                return num
        except ValueError:
            print("Invalid input. Please enter an integer.")

def factorial(n):
    if n == 0 or n == 1:  # base case
        return 1
    else:
        return n * factorial(n-1)  # recursive call

def main():
    num = get_number()
    fact = factorial(num)
    print(f"The factorial of {num} is {fact}")

if __name__ == "__main__":
    main()
```

```python
from functools import reduce

def get_number():
    while True:
        try:
            num = int(input("Enter a non-negative integer: "))
            if num < 0:
                print("Factorial is not defined for negative numbers. Try again!")
            else:
                return num
        except ValueError:
            print("Invalid input. Please enter an integer.")

def factorial(n):
    return reduce(lambda x, y: x*y, range(1, n+1), 1)

def main():
    num = get_number()
    fact = factorial(num)
    print(f"The factorial of {num} is {fact}")

if __name__ == "__main__":
    main()
```

```python
student_names = ['Alice', 'Bob', 'Charile']
assignment1_scores = [85, 92, 78]
assignment2_scores = [90, 88, 80]
student_records = zip(student_names, assignment1_scores, assignment2_scores)
student_data_dict = {name: (assign1, assign2) for name, assign1, assign2 in student_records}

print("Student Records Dictionary:")
print(student_data_dict)

student_total_scores = {}
for name, scores in student_data_dict.items():
    student_total_scores[name] = sum(scores)

print("\nStudent Total Scores Dictionary:")
print(student_total_scores)
```

```python
def are_anagrams(word1, word2):
    # Convert both words to lowercase and remove spaces
    word1 = word1.lower().replace(" ", "")
    word2 = word2.lower().replace(" ", "")

    # Check if the sorted versions of the words are equal
    return sorted(word1) == sorted(word2)

# Get input from the user
input_word1 = input("Enter the first word: ")
input_word2 = input("Enter the second word: ")

# Check if they are anagrams and print the result
if are_anagrams(input_word1, input_word2):
    print(f"'{input_word1}' and '{input_word2}' are anagrams.")
else:
    print(f"'{input_word1}' and '{input_word2}' are not anagrams.")
```
