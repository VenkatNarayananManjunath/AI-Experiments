```python
# Python Implementation
import numpy as np
import random
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Load dataset
data = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(
    data.data, data.target, test_size=0.2, random_state=42
)

# Genetic Algorithm parameters
POPULATION_SIZE = 10
N_GENERATIONS = 20
MUTATION_PROB = 0.2
PARAM_BOUNDS = {'C': (0.001, 10)}  # Hyperparameter bounds

# Initialize population
def init_population(size):
    return [{'C': random.uniform(*PARAM_BOUNDS['C'])} for _ in range(size)]

# Fitness function using validation accuracy
def fitness(individual):
    # Increased max_iter to address ConvergenceWarning
    model = LogisticRegression(C=individual['C'], max_iter=1000)
    model.fit(X_train, y_train)
    predictions = model.predict(X_test)
    return accuracy_score(y_test, predictions)

# Selection: Top 50% individuals
def selection(pop, scores):
    sorted_pop = [p for _, p in sorted(zip(scores, pop), key=lambda x: x[0], reverse=True)]
    return sorted_pop[:len(pop)//2]

# Crossover: Average C parameter
def crossover(parent1, parent2):
    child = {'C': (parent1['C'] + parent2['C']) / 2}
    return child

# Mutation: Random small change in C
def mutate(individual):
    if random.random() < MUTATION_PROB:
        individual['C'] *= random.uniform(0.8, 1.2)
        # Ensure C stays within bounds
        individual['C'] = max(min(individual['C'], PARAM_BOUNDS['C'][1]), PARAM_BOUNDS['C'][0])
    return individual

# Run Genetic Algorithm
population = init_population(POPULATION_SIZE)

gen_best_scores = [] # To store best scores per generation
gen_best_params = [] # To store best parameters per generation

for gen in range(N_GENERATIONS):
    scores = [fitness(ind) for ind in population]
    population = selection(population, scores)
    children = []
    while len(children) < POPULATION_SIZE - len(population):
        parents = random.sample(population, 2)
        child = crossover(parents[0], parents[1])
        child = mutate(child)
        children.append(child)
    population.extend(children)
    best_score = max(scores)
    best_individual = population[scores.index(best_score)]
    gen_best_scores.append(best_score)
    gen_best_params.append(best_individual)
    print(f"Generation {gen+1}, Best Accuracy: {best_score:.4f}, Best Params: {best_individual}")

# Final best model
final_best_score_overall = max(gen_best_scores)
final_best_individual_overall = gen_best_params[gen_best_scores.index(final_best_score_overall)]

final_model = LogisticRegression(C=final_best_individual_overall['C'], max_iter=1000)
final_model.fit(X_train, y_train)
final_acc = accuracy_score(y_test, final_model.predict(X_test))
print(f"Final Best Accuracy: {final_acc:.4f}, Best C: {final_best_individual_overall['C']:.4f}")

import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(range(1, N_GENERATIONS + 1), gen_best_scores, marker='o')
plt.title('Best Accuracy Over Generations')
plt.xlabel('Generation')
plt.ylabel('Best Accuracy')
plt.grid(True)
plt.xticks(range(1, N_GENERATIONS + 1))
plt.show()
```
