```python
import tensorflow as tf
x = tf.constant([2.0, 9.8, 0.1]) # Subtract max for numerical stability
x_max = tf.reduce_max(x) # Exponentiation
exp_x = tf.exp(x - x_max)
softmax = exp_x / tf.reduce_sum(exp_x) # Normalization
print("Softmax output:", softmax)
```

```python
import tensorflow as tf
x = tf.constant([9.7, 9.6, 7.3, 3.2])
x_min = tf.reduce_min(x)
x_max = tf.reduce_max(x)
normalized = (x - x_min) / (x_max - x_min)
print("Normalized tensor:", normalized)
```

```python
import torch
x = torch.tensor(8.0, requires_grad=True) # Create a tensor with gradient tracking enabled
y = x ** 2 # Tensor operation
y.backward() # Compute gradient
print("Gradient dy/dx:", x.grad) # Print gradient
```

```python
import torch
A = torch.tensor([1.7, 2.8, 3.7]) # Input tensors
B = torch.tensor([4.7, 5.5, 6.5])
dot_product = torch.sum(A * B) # Dot product
norm_A = torch.sqrt(torch.sum(A * A)) # Norms
norm_B = torch.sqrt(torch.sum(B * B))
cosine_similarity = dot_product / (norm_A * norm_B) # Cosine similarity
print(cosine_similarity)
```
