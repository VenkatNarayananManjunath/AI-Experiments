```python
# Install dependencies
!pip install transformers==4.40.0 torch torchvision torchaudio pillow

# Import libraries
from transformers import BlipProcessor, BlipForConditionalGeneration
from PIL import Image
from google.colab import files
import torch

# Load BLIP model + processor
processor = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-base")
model = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-base")

# Upload multiple images
print(" Please upload your images...")
uploaded = files.upload()

# Loop through all uploaded images and summarize activities
for filename in uploaded.keys():
    try:
        image = Image.open(filename).convert("RGB")
        inputs = processor(images=image, return_tensors="pt")
        out = model.generate(**inputs)
        caption = processor.decode(out[0], skip_special_tokens=True)
        print(f"{filename} → 📝 Activity Summary: {caption}")
    except Exception as e:
        print(f"⚠️ Error processing {filename}: {e}")
```
