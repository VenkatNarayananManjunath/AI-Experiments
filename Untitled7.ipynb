```python
import tensorflow as tf

# Create two random (2×3) tensors
tensor1 = tf.random.uniform(shape=(2, 3), minval=0, maxval=10)
tensor2 = tf.random.uniform(shape=(2, 3), minval=0, maxval=10)

print("Tensor 1:\n", tensor1)
print("\nTensor 2:\n", tensor2)

# Element-wise addition
result = tf.add(tensor1, tensor2)

print("\nElement-wise Addition Result:\n", result)
```

```python
import tensorflow as tf

print("Enter values for Tensor A (2×3):")
A = []
for i in range(2):
    row = list(map(float, input(f"Row {i+1}: ").split()))
    A.append(row)

print("Enter values for Tensor B (2×3):")
B = []
for i in range(2):
    row = list(map(float, input(f"Row {i+1}: ").split()))
    B.append(row)

# Convert to TensorFlow tensors
A = tf.constant(A, dtype=tf.float32)
B = tf.constant(B, dtype=tf.float32)

# Element-wise addition
C = tf.add(A, B)

print("\nTensor A:\n", A)
print("\nTensor B:\n", B)
print("\nElement-wise Addition Result:\n", C)
```

```python
# Take input for Matrix A
print("Enter values for Matrix A (2×3):")
A = []
for i in range(2):
    while True:
        row_str = input(f"Row {i+1} (enter 3 numbers separated by spaces): ")
        row = list(map(float, row_str.split()))
        if len(row) == 3:
            A.append(row)
            break
        else:
            print(f"Invalid input. Please enter exactly 3 numbers for Row {i+1}.")

# Take input for Matrix B
print("Enter values for Matrix B (2×3):")
B = []
for i in range(2):
    while True:
        row_str = input(f"Row {i+1} (enter 3 numbers separated by spaces): ")
        row = list(map(float, row_str.split()))
        if len(row) == 3:
            B.append(row)
            break
        else:
            print(f"Invalid input. Please enter exactly 3 numbers for Row {i+1}.")

# Initialize result matrix
C = [[0 for _ in range(3)] for _ in range(2)]

# Element-wise addition
for i in range(2):
    for j in range(3):
        C[i][j] = A[i][j] + B[i][j]

# Display results
print("\nMatrix A:")
for row in A:
    print(row)

print("\nMatrix B:")
for row in B:
    print(row)

print("\nElement-wise Addition Result:")
for row in C:
    print(row)
```
